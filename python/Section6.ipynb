```python
import planckStyle as s
g4 = s.getSinglePlotter(chain_dir=['../main/'])

pars = [[],['omegak'],['w'],['omegak','w'],['w','wa'],['omegak','w','wa'],['mnu'],['mnu','w']]
parsLabel = ['\Lambda \mathrm{CDM}','\mathrm{o}\Lambda\mathrm{CDM}','\mathrm{wCDM}','\mathrm{owCDM}','w_0 w_a\mathrm{CDM}'
            ,'\mathrm{o}w_0 w_a\mathrm{CDM}',r'm_\nu\Lambda \mathrm{CDM}',r'm_\nu\mathrm{wCDM}']

toppars = ['omegal','H0','sigma8','omegak','w','wa','mnu']

datatag = ['CMBLens_BAORSD_DES_SN']
lines = []
heading = ''
for i, par in enumerate(pars):
    for j, dat in enumerate(datatag):
        print(par,dat)
        samples = g4.getSamples('_'.join(par), dat)
        samples.paramNames.setLabelsAndDerivedFromParamNames(g4.settings.param_names_for_labels)
        line = ''
        if j == 0: line = str(parsLabel[i])
        latex = samples.getLatex(toppars, limit=1, err_sig_figs=2)
        if 'mnu' in par: 
            stats = samples.getMargeStats()
            latex[1][6]='<'+'%s'%float('%.3g' % stats.parWithName('mnu').limits[1].upper)
        if j == 0 and i<len(toppars):
            heading += ' & '+ str(latex[0][i])
            print(line)

        line += r' & ' + ' & '.join([str(i or '-') for i in latex[1]]) + '\\\\'
        if j == len(datatag)-1: line += '\n \\hline'
        lines.append(line)
        
savefile = '../python/tables_final/Table_full.tex'
outF = open(savefile, "w")
outF.writelines('\\begin{tabular}{|C|C|C|C|C|C|C|C|C|C|}')
outF.writelines('\n \hline \hline \n'+ heading + '\\\\\n\\hline\n' + '\n'.join(lines) + '\n')
outF.writelines('\\end{tabular}\n ')
outF.close()
```

```python
import planckStyle as s
g = s.getSubplotPlotter()

pars = ['w','wa']

datatag = ['CMBLens_BAORSD','CMBLens_SN','CMBLens_BAORSD_DES_SN']
dataLabel = ['\\mathrm{'+i.replace('_',' + ')+'}' for i in datatag]

lines = []
heading = '\mathrm{Data}& w_0 &w_a & \mathrm{FoM}&z_p '
for j, dat in enumerate(datatag):
    print(pars,dat)
    samples = g.getSamples('_'.join(pars), dat) 
    samples.paramNames.setLabelsAndDerivedFromParamNames(g.settings.param_names_for_labels)
    stats = samples.getMargeStats()
    FoM = 1./np.sqrt(np.linalg.det(samples.cov(pars=pars)))
    ap = 1+samples.cov(pars=pars)[0,1]/samples.cov(pars=pars)[1,1]
    zp=1/ap-1
    print('zp',zp)
    print('FoM',FoM)
    p = samples.getParams()
    samples.addDerived(p.w+(1-ap)*p.wa, name='wp', label='w_p')
    
    derived = p.w+(1-ap)*p.wa

    print('mean = %s, err = %s'%(samples.mean(derived), samples.std(derived)))
    latex = samples.getLatex(pars, limit=1, err_sig_figs=2)
    lines.append(dataLabel[j]+r' & ' + ' & '.join([str(i or '-') for i in latex[1]]) +
                 ' & '+'%s' % float('%.4g' % FoM)+r' & '+'%s' % float('%.2g' % zp)+'\\\\')
   
lines.append('\hline')
print(lines)
savefile = '../python/tables_final/Table_w0waCDM.tex'
outF = open(savefile, "w")
outF.writelines('\\begin{tabular}{|L|C|C|C|C|} \n')
outF.writelines('\n \hline')
outF.writelines('\n \hline')
outF.writelines(heading + '\\\\\n\\hline\n' + '\n'.join(lines) + '\n')
outF.writelines('\end{tabular}\n ')
outF.close()
```
