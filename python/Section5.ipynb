```python
g4 = plots.getSinglePlotter(chain_dir=['../../mgcosmomc2020/chains/'],rc_sizes=True)
rootname='../../mgcosmomc2020/chains/Mu_SDSS_final_RSD_submit'
samples = loadMCSamples(rootname)
stats = samples.getMargeStats()
print(stats.parWithName('mu0').mean)
print('R-1',samples.getGelmanRubin())
roots = ['MuSigma_final_WL_lens_v5','MuSigma_SDSS_final_RSD_WL_lens_submit']
plt.fill_between([-1,1],stats.parWithName('mu0').mean-stats.parWithName('mu0').err
                 ,y2=stats.parWithName('mu0').mean+stats.parWithName('mu0').err
                 ,color='#006FED',alpha=0.85)
plt.fill_between([-1,1],stats.parWithName('mu0').mean-2*stats.parWithName('mu0').err
                 ,y2=stats.parWithName('mu0').mean+2*stats.parWithName('mu0').err
                 ,color='#006FED',alpha=0.3)
g4.plot_2d(roots,['sigma0','mu0'], filled=True,colors=['#E03424','grey'])
#g.add_y_bands(stats.parWithName('mu0').mean,stats.parWithName('mu0').err,color='#006FED')
g4.add_2d_contours(roots[0],'sigma0','mu0',filled=True,color='#E03424')
g4.add_2d_contours(roots[1],'sigma0','mu0',filled=True,color='grey')
#g.add_legend(['WL+CMB lens','all'],legend_loc='upper right')

red_patch = mpatches.Patch(color='#E03424', label='WL+CMB lens')
blue_patch = mpatches.Patch(color='#006FED', label='RSD')
grey_patch = mpatches.Patch(color='grey', label='combined')

plt.legend(handles=[blue_patch,red_patch,grey_patch],fontsize=15,loc='lower left')
g4.add_text('MG',0.95,0.1,fontsize=18)

#g.add_y_marker(0)
#g.add_x_marker(0)
plt.xlim(-0.6,.4)
plt.ylim(-1.0,0.7)
g4.export('plots_final/Sigma0_mu0_RSD_WL_lens.pdf')
g4.export('plots_final/Sigma0_mu0_RSD_WL_lens.png')
for r in roots:
    samples = loadMCSamples('../../mgcosmomc2020/chains/'+r)
    stats = samples.getMargeStats()
    print('R-1',samples.getGelmanRubin())
```

```python
import planckStyle as s

g5 = s.getSinglePlotter(chain_dir=['../main/'])

pars = [['omegak'],['w']]
parsLabel = ['\mathrm{o}\Lambda\mathrm{CDM}','\mathrm{wCDM}']

toppars = ['omegam', 'omegal','sigma8','omegak','w']

datatag = ['CMB','CMB_RSD','CMB_WL','CMB_allGrowth']
dataLabel = ['\mathrm{CMB \ T}$\&$\mathrm{P}','\mathrm{CMB \ T}$\&$\mathrm{P+RSD}'
             ,'\mathrm{CMB \ T}$\&$\mathrm{P+WL}','\mathrm{CMB \ T}$\&$\mathrm{P+RSD+WL+CMB \ lens}']
lines = []
heading = '& & \Omega_m & \Omega_\mathrm{DE} & \sigma_8 & \Omega_k & w'
for i, par in enumerate(pars):
    for j, dat in enumerate(datatag):
        print(par,dat)
        samples = g5.getSamples('_'.join(par), dat)
        samples.paramNames.setLabelsAndDerivedFromParamNames(g5.settings.param_names_for_labels)
        varied = [samples.paramNames.parWithName(p).getLabel() for p in par]
        line = ''
        if j == 0: line = '\\multirow{'+str(len(datatag))+'}{*}{$'+str(parsLabel[i])+'$}'
        line += ' & ' + dataLabel[j]
        latex = samples.getLatex(toppars, limit=1, err_sig_figs=2)

        line += r' & ' + ' & '.join([str(i or '-') for i in latex[1]]) + '\\\\'
        if j == len(datatag)-1: line += '\n \\hline'
        lines.append(line)
        
savefile = '../python/tables_final/Table_ext_RSD.tex'
outF = open(savefile, "w")
outF.writelines('\\begin{tabular}{|C|C|C|C|C|C|C|}')
outF.writelines('\n \hline \hline'+ heading + '\\\\\n\\hline\n' + '\n'.join(lines) + '\n')
outF.writelines('\\end{tabular}\n ')
outF.close()
```
