```python
import planckStyle as s
g5 = s.getSinglePlotter(chain_dir=['../main/'])

pars = [[],['omegak'],['w'],['mnu']]
parsLabel = ['\Lambda \mathrm{CDM}','\mathrm{o}\Lambda\mathrm{CDM}','\mathrm{wCDM}',r'\nu \Lambda \mathrm{CDM}']

toppars = ['omegal','H0', 'omegak','w','mnu']

datatag = ['BAOonly','CMB','CMB_BAO','CMB_SN','CMB_BAO_SN']
dataLabel = ['\\mathrm{'+i.replace('_',' + ')+'}' for i in datatag]
dataLabel = [i.replace('CMB','CMB \ T\&P')for i in dataLabel]
dataLabel = [i.replace('BAOonly','BAO')for i in dataLabel]
lines = []
heading = '& &  \Omega_\mathrm{DE} & H_0 \mathrm{[km/s/Mpc]} & \Omega_k & w &\\Sigma m_\\nu\\,[\\mathrm{eV}]'
for i, par in enumerate(pars):
    for j, dat in enumerate(datatag):
        if (j!=0 or i!=3):
            samples = g5.getSamples('_'.join(par), dat)
            samples.paramNames.setLabelsAndDerivedFromParamNames(g5.settings.param_names_for_labels)
            varied = [samples.paramNames.parWithName(p).getLabel() for p in par]
            line = ''
            if j == 1: line = '\\multirow{'+str(len(datatag))+'}{*}{$'+str(parsLabel[i])+'$}'
            line += ' & ' + dataLabel[j]
            latex = samples.getLatex(toppars, limit=1, err_sig_figs=2)
            if (j==1 and 'w' in par) or (j==0): latex[1][1]='-'
            if 'mnu' in par: 
                stats = samples.getMargeStats()
                latex[1][4]='<'+'%s'%float('%.3g' % stats.parWithName('mnu').limits[1].upper)+'\ (95\%)'

            line += r' & ' + ' & '.join([str(s or '-') for s in latex[1]]) + '\\\\'
            if j == len(datatag)-1: line += '\n \\hline'

            lines.append(line)
  
savefile = '../python/tables_final/Table_LCDM_BAO.tex'
outF = open(savefile, "w")
outF.writelines('\\begin{tabular}{|L|L|C|C|C|C|C|}')
outF.writelines('\hline \hline \n'+ heading + '\\\\\n\\hline\n' + '\n'.join(lines) + '\n')
outF.writelines('\\end{tabular}\n ')
outF.close()
```
